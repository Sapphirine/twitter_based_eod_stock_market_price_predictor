```python
import pandas as pd
import os
```

```python
counter = 0
amazon = pd.DataFrame()
apple = pd.DataFrame()
att = pd.DataFrame()
disney = pd.DataFrame()
microsoft = pd.DataFrame()
netflix = pd.DataFrame()
verizon = pd.DataFrame()
walmart = pd.DataFrame()

for root, dirs, files in os.walk('Extra data June/'):
    if counter:
        print 'Initializing dataframe and adding data...'
        amazon = pd.concat([amazon,pd.DataFrame(pd.read_csv('%s/%s' % (root, files[0]), names = ['Twitter', 'Date']), columns=['Twitter', 'Date'])], ignore_index=True)
        apple = pd.concat([apple,pd.DataFrame(pd.read_csv('%s/%s' % (root, files[1]), names = ['Twitter', 'Date']), columns=['Twitter', 'Date'])], ignore_index=True)
        att = pd.concat([att,pd.DataFrame(pd.read_csv('%s/%s' % (root, files[2]), names = ['Twitter', 'Date']), columns=['Twitter', 'Date'])], ignore_index=True)
        disney = pd.concat([disney,pd.DataFrame(pd.read_csv('%s/%s' % (root, files[3]), names = ['Twitter', 'Date']), columns=['Twitter', 'Date'])], ignore_index=True)
        microsoft = pd.concat([microsoft,pd.DataFrame(pd.read_csv('%s/%s' % (root, files[4]), names = ['Twitter', 'Date']), columns=['Twitter', 'Date'])], ignore_index=True)
        netflix = pd.concat([netflix,pd.DataFrame(pd.read_csv('%s/%s' % (root, files[5]), names = ['Twitter', 'Date']), columns=['Twitter', 'Date'])], ignore_index=True)
        verizon = pd.concat([verizon,pd.DataFrame(pd.read_csv('%s/%s' % (root, files[6]), names = ['Twitter', 'Date']), columns=['Twitter', 'Date'])], ignore_index=True)
        walmart = pd.concat([walmart,pd.DataFrame(pd.read_csv('%s/%s' % (root, files[7]), names = ['Twitter', 'Date']), columns=['Twitter', 'Date'])], ignore_index=True)
    counter+=1
data_list = [amazon,apple,att,disney,microsoft,netflix,verizon,walmart]
company_list=[]
for df in data_list:
    print 'Building company list...'
    company_list.append(pd.DataFrame(df.groupby('Date')['Twitter'].apply(lambda x:x.sum())).reset_index())
for root, dirs, files in os.walk('Twitter Results/PivotedResults/'):
    print 'Joining Pivoted data...'
    company_list[0] = pd.concat([company_list[0],pd.DataFrame(pd.read_csv('%s/%s' % (root, files[0]), names = ['Date','Twitter']), columns=['Date','Twitter'])], ignore_index=True)
    company_list[1] = pd.concat([company_list[1],pd.DataFrame(pd.read_csv('%s/%s' % (root, files[1]), names = ['Date','Twitter']), columns=['Date','Twitter'])], ignore_index=True)
    company_list[2] = pd.concat([company_list[2],pd.DataFrame(pd.read_csv('%s/%s' % (root, files[2]), names = ['Date','Twitter']), columns=['Date','Twitter'])], ignore_index=True)
    company_list[3] = pd.concat([company_list[3],pd.DataFrame(pd.read_csv('%s/%s' % (root, files[3]), names = ['Date','Twitter']), columns=['Date','Twitter'])], ignore_index=True)
    company_list[4] = pd.concat([company_list[4],pd.DataFrame(pd.read_csv('%s/%s' % (root, files[4]), names = ['Date','Twitter']), columns=['Date','Twitter'])], ignore_index=True)
    company_list[5] = pd.concat([company_list[5],pd.DataFrame(pd.read_csv('%s/%s' % (root, files[5]), names = ['Date','Twitter']), columns=['Date','Twitter'])], ignore_index=True)
    company_list[6] = pd.concat([company_list[6],pd.DataFrame(pd.read_csv('%s/%s' % (root, files[6]), names = ['Date','Twitter']), columns=['Date','Twitter'])], ignore_index=True)
    company_list[7] = pd.concat([company_list[7],pd.DataFrame(pd.read_csv('%s/%s' % (root, files[7]), names = ['Date','Twitter']), columns=['Date','Twitter'])], ignore_index=True)

def reparse_date(df):
    for item in df.iterrows():
        date = item[1][0].split()[0].split('-')
        date = '%s_%s_%s' % (date[1],date[2],date[0][-2:])
        df.loc[item[0],'Date'] = date
    return df
quandl_list=[]
for root, dirs, files in os.walk('Quandl Prices/'):
    for i in range(len(files)):
        print 'Building Quandl list...'
        test = pd.DataFrame.from_csv('%s/%s' % (root, files[i]))
        quandl_list.append(reparse_date(test))
        
joined_list = []
for i in range(len(quandl_list)):
    print 'Resetting index and joining table...'
    quandl_list[i]=quandl_list[i].reset_index(drop=True)
    company_list[i] = company_list[i].reset_index(drop=True)
    joined_list.append(pd.DataFrame(company_list[i].set_index('Date').join(quandl_list[i].set_index('Date'))))

```

```python
from sklearn.feature_extraction import DictVectorizer
from sklearn.pipeline import Pipeline
from sklearn import pipeline
from sklearn.linear_model import LinearRegression
from sklearn.feature_extraction.text import TfidfVectorizer


dataX = joined_list[0]['Twitter']
vectorizer = TfidfVectorizer()
#X = vectorizer.fit_transform(dataX)

linreg = LinearRegression(fit_intercept=True) 
att_pipe = pipeline.Pipeline([
  ('tfidf', vectorizer),
  ('linreg', linreg)
  ])
att_pipe.fit(dataX,joined_list[0]['Movement'])
```

```python
att_pipe.predict(['''purge'''])
```

```python
import dill
name_list = ['amazon','apple','att','disney','microsoft','netflix','verizon','walmart']
for i in range(len(joined_list)):
    att_pipe.fit(joined_list[i]['Twitter'],joined_list[i]['Movement'])
    dill.dump(att_pipe, open("%s.pkl" % name_list[i], "w"))
```
